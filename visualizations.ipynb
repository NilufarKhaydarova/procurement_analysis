```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import hvplot as hv
from PIL import Image
import plotly.express as px
```

```python
vendor_terr_names = pd.read_csv('data/vendor_terr_names.csv', sep=',\t', names= ['id', 'name'])
vendor_terr_names.info()
vendor_terr_names.head(10)
```

```python
resultat = pd.read_csv('data/resultat_method.csv', sep=',')
resultat.info()
resultat.head(10)
```

```python
#replace all missing values in vendor_terr with 1111
resultat['vendor_terr'].fillna(1111, inplace=True)
```

```python
resultat.vendor_terr = resultat.vendor_terr.astype('int64')
```

```python
rel_db = pd.read_csv('data/resultat_method_specifications.csv', sep=',')
rel_db.info()
rel_db.head(10)
```

```python
#join resultat_method and rel_db on resultat.id and rel_db.resultat_method_id
resultat = resultat.merge(rel_db, left_on='id', right_on='resultat_method_id')
resultat.info()
```

```python
specs = pd.read_csv('data/specifications.csv', sep=',')
specs.info()
specs.head(10)
```

```python
#join resultat and specifications on resultat.specification_id and specifications.id
resultat = resultat.merge(specs, left_on='specifications_id', right_on='id')
resultat.info()
resultat.head(10)
```

```python
resultat = resultat.drop(['id_x', 'id_y', 'specifications_id',  'resultat_method_id'], axis=1)
```

```python
#only keep rows where proc_id = 6
resultat = resultat[resultat['proc_id'] == 6]
resultat.info()
```

```python
contract = pd.read_csv('data/contract_info.csv', sep=',')
contract.info()
contract.head(10)
```

```python
#drop duplicates response_id keep last
contract.drop_duplicates(subset='response_id', keep='last', inplace=True)
```

```python
contract = contract[contract.state == 2]
```

```python
contract = contract.drop_duplicates('lot_id')
```

```python
contract.request_id.head()
```

```python
contract.info()
```

```python
resultat = resultat[resultat.proc_id == 6]
```

```python
#is lot_id in contract same as lot_id in resultat
resultat['lot_id'].isin(contract['lot_id']).value_counts()
```

```python
#only keep rows where lot_id is in contract
resultat = resultat[resultat['lot_id'].isin(contract['lot_id'])]
```

```python
resultat.info()
```

```python
df = pd.merge(resultat, contract, on='lot_id', how='left')
```

```python
df.info()
df.head(10)
```

```python
#is request_id in contract same as request_id in resultat
df['request_id_x'].isin(df['request_id_y']).value_counts()
```

```python
#keep only rows where request_id_x is in request_id_y
df = df[df['request_id_x'].isin(df['request_id_y'])]
```

```python
df = df.drop('request_id_y', axis=1)
```

```python
df['request_id'] = df['request_id_x']
```

```python
#extract month from contract_dat and plot contract_num
df['contract_dat'] = pd.to_datetime(df['contract_dat'], errors='coerce')
```

```python
df['contract_num'] = df['contract_num_x']
df = df.drop(['contract_num_x', 'contract_num_y'], axis=1)
```

```python
df = df.drop(['version_x', 'version_y', 'errmsg', 'state', 'request_id_x', 'method_name_x', 'method_name_y', 'gen_id_x', 'gen_id_y', 'kls_x', 'kls_y'], axis=1)
```

```python
df.info()
```

```python
months = pd.to_datetime(df['contract_dat'], format='%m').dt.month_name(locale='ru_RU').str.slice(stop=3)

#plot contract_num by month
plt.figure(figsize=(10, 5))
sns.countplot(x=months, data=df.contract_num)
plt.xticks(rotation=90)
#show numbers on top of bars
for p in plt.gca().patches:
    plt.gca().annotate('{:.0f}'.format(p.get_height()), (p.get_x()+p.get_width()/2., p.get_height()), 
                       ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.title('Количество контрактов по месяцам')
```
